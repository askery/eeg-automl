```python
from tpot import TPOTClassifier
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# load raw data
pref = ['C4_1_','C4_5_','C4_6_', 'C4_8_', 'C4_10_']
#pref = ['C4_10_']
df_d = {}
for i in range(len(pref)):
    fname   = pref[i] + 'rawdata_open.csv'
    df_d[i] = pd.read_csv(fname)
#df = pd.read_csv('C4_10_features.csv')
#df_d[4]  
```

```python
df1 = pd.concat(df_d.values(), ignore_index=True)
#ii = 3
#df = df_d[ii]  
df1
```

```python
# load features
df_d2 = {}
for i in range(len(pref)):
    fname   = pref[i] + 'features.csv'
    df_d2[i] = pd.read_csv(fname, decimal=',')
#df = pd.read_csv('C4_10_features.csv')
#df_d[4]  
```

```python
df2 = pd.concat(df_d2.values(), ignore_index=True)
#ii = 3
#df = df_d[ii]  
df2
```

```python
df = pd.concat([df1.iloc[:,0:df1.shape[1] -1],df2.iloc[:,0:df2.shape[1] -1], df1.iloc[: , -1:]], ignore_index=True,  axis=1)
```

```python
# inspect
df
```

```python
# inspect types 
df.dtypes
```

```python
df.shape[1]
```

```python
X = df.iloc[:,0:df.shape[1] -1].values
#
#scaler = MinMaxScaler()
#X  = scaler.fit_transform(X.T).T
#
y = df.iloc[: , -1:].values
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    train_size=0.8, random_state=18, stratify=y)
```

```python
# TPOT cfg
tpot = TPOTClassifier(verbosity=2, 
                      scoring="accuracy", 
                      random_state=11,  
                      generations=15, 
                      population_size=15,
                      early_stop=5)
```

```python
%%time
# fit
tpot.fit(X_train, y_train)
```

```python
# TPOT Predict
y_pred = tpot.predict(X_test)
```

```python
# Results
print(classification_report(y_test, y_pred))
```

```python
# sanity check (accuracy)
acc = tpot.score(X_test, y_test)
acc
```

```python
# confusion matrix
cm  = confusion_matrix(y_test, y_pred)
cm
```
